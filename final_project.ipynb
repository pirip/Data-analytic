#Average earning per job title


```python
avgearn=df.groupby('job_title').mean()['salary_in_usd'].reset_index()
avgearn=avgearn.sort_values('salary_in_usd',ascending=False).head(10)
avgearn
```

#Displaying in a bar chart



```python
avgearn.plot(kind='bar',x='job_title',y='salary_in_usd',title='Average earning per job title',figsize=(10,5))
```

#Average pay by region


```python
payus=df[df.company_location=='US']
payus=payus.groupby('job_title').mean()['salary_in_usd'].reset_index()
payus=payus.sort_values('salary_in_usd',ascending=False).head(10)
payus
```

#Comparing US pay with rest of region


```python
payout=df[df.company_location!='US']
payout=payout.groupby('job_title').mean()['salary_in_usd'].reset_index()
payout=payout.sort_values('salary_in_usd',ascending=False).head(10)
payout
```

```python
compare = pd.merge(payus,payout[['job_title','salary_in_usd']],on='job_title', how ='left').dropna().head(10)
compare
```

#Graphical representation:



```python
compare.plot(kind='bar',figsize=(10,5),x='job_title',title='Comparison of average pay per job title for US companies to other parts')
```

#salary with company size


```python
compsizeUS=df[df.company_location=='US']
compsizeUS=compsizeUS.groupby('company_size').mean()['salary_in_usd'].reset_index()
compsizeUS=compsizeUS.sort_values('salary_in_usd',ascending=False)
compsizeUS
```

#for company outside of us


```python
compsizeNOTUS=df[df.company_location!='US']
compsizeNOTUS=compsizeUS.groupby('company_size').mean()['salary_in_usd'].reset_index()
compsizeNOTUS=compsizeNOTUS.sort_values('salary_in_usd',ascending=False)
compsizeNOTUS
```

#make a comparision


```python
sizecomp = pd.merge(compsizeUS,compsizeNOTUS[['company_size','salary_in_usd']],on='company_size', how='left')
sizecomp
```

```python
sizecomp.plot(kind='bar',figsize=(10,5),x='company_size')
```
